```python
import mysql.connector
import csv
import os

# Database connection parameters
conn = mysql.connector.connect(
    host='localhost',
    user='root',            
    password=os.getenv("MYSQL_PASS"),
    database='finances'  
)
 
cursor = conn.cursor()

# Step 1: Create the table
cursor.execute("""
CREATE TABLE IF NOT EXISTS pocketsmith_search (
    Date DATE,
    Merchant VARCHAR(255),
    Merchant_Changed_From VARCHAR(255),
    Amount DECIMAL(15,2),
    Currency VARCHAR(8),
    Amount_in_base_currency DECIMAL(15,2),
    Base_currency VARCHAR(8),
    Transaction_Type VARCHAR(32),
    Account VARCHAR(255),
    Closing_Balance DECIMAL(15,2),
    Category VARCHAR(255),
    Parent_Categories VARCHAR(255),
    Labels VARCHAR(255),
    Memo VARCHAR(255),
    Note VARCHAR(255),
    ID BIGINT
)
""")

# Step 2: Prepare the insert statement
insert_stmt = """
INSERT INTO pocketsmith_search VALUES (
    %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s
)
"""

# Step 3: Open the CSV and load the data
with open('pocketsmith-search.csv', 'r', encoding='utf-8') as f:
    reader = csv.reader(f)
    header = next(reader)  # Skip header row
    for row in reader:
        # Convert empty strings to None if your DB should have NULLs
        row = [None if x == "" else x for x in row]
        cursor.execute(insert_stmt, row)

# Commit and clean up
conn.commit()
cursor.close()
conn.close()
```
